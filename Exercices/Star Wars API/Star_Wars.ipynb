Trier par classification et récupération uniquement de name et classification

```python
import requests
import pandas as pd

def species_data():
    next_page = "https://swapi.dev/api/species/"
    all_results = []

    while next_page:
        response = requests.get(next_page)
        if response.status_code == 200:
            data = response.json()

            all_results.extend(data['results'])

            next_page = data['next']
        else:
            print("Erreur lors de la récupération des données.")
            return None

    df = pd.DataFrame(all_results)
    return df

species_df = species_data()

if species_df is not None:
    selected_columns = ['name', 'classification']
    species_df = species_df[selected_columns].sort_values(by='classification')
    display(species_df)
```

Groupé les classifications dans des dataframes dédiés

```python
import requests
import pandas as pd

def swapi_species_data():
    next_page = "https://swapi.dev/api/species/"
    all_results = []

    while next_page:
        response = requests.get(next_page)
        if response.status_code == 200:
            data = response.json()

            all_results.extend(data['results'])

            next_page = data['next']
        else:
            print("Erreur lors de la récupération des données.")
            return None

    df = pd.DataFrame(all_results)
    return df

def split_by_classification(df):
    # Vérifier si le DataFrame n'est pas vide
    if not df.empty:
        # Créer un dictionnaire pour stocker les DataFrames par classification
        dfs_by_classification = {}
        
        # Grouper le DataFrame par classification
        grouped = df.groupby('classification')
        
        # Parcourir chaque groupe et stocker les résultats dans le dictionnaire
        for classification, group_df in grouped:
            dfs_by_classification[classification] = group_df
        
        return dfs_by_classification
    else:
        print("Le DataFrame est vide.")
        return None

swapi_species_df = swapi_species_data()

if swapi_species_df is not None:
    # Sélectionner uniquement les colonnes 'name' et 'classification'
    selected_columns = ['name', 'classification']
    # Trier le DataFrame par nom
    swapi_species_df = swapi_species_df[selected_columns].sort_values(by='name')
    
    # Diviser le DataFrame par classification
    dfs_by_classification = split_by_classification(swapi_species_df)
    
    if dfs_by_classification is not None:
        # Afficher les DataFrames par classification
        for classification, df_classification in dfs_by_classification.items():
            print(f"Classification: {classification}")
            display(df_classification)
            print()
```
